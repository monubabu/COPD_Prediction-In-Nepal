### Spilt the data into training and testing

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Load the data
path = r'M:\Data Science\COPD_Prediction\EDA\engineered_COPD_data.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.columns
```

```python
# Define the features and the target variable
X = df.drop(columns =['COPD_Diagnosis'])
y = df['COPD_Diagnosis']
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Model Training
- Based on the data, it is a binary classification problem set as we are prediction if someone has COPD or not, hence we are goint to use the models:
   - Logistic Regression
   - Decision Trees
   - Random Forest

```python
# Import the necessary Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import pickle
```

```python
# Initialize the models
models = {
    'Logistic Regression' : LogisticRegression(max_iter=1000),
    'Decision Tree' : DecisionTreeClassifier(),
    'Random Forest' : RandomForestClassifier()
}
```

```python
# Train the models
for name, model in models.items():
    model.fit(X_train, y_train)

    # Save the models
    with open(f'{name.replace(" ","_")}.pkl', 'wb') as f:
        pickle.dump(model, f)
        print(f"{name} model trained and saved.")

print('Model training completed')
```

## Evaluate the models
- Accuracy, Precison, Recall, F1_Score

```python
for name, model in models.items():
    y_prep = model.predict(X_test)
    print(f"\n{name} Evaluation: ")
    print(classification_report(y_test, y_prep))
```

### Model Refinement

```python
# import the libraries
from sklearn.model_selection import GridSearchCV
```

```python
# Define the parameter grid for Random forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
```

```python
# Initialize the GridSearchCV
grid_search = GridSearchCV(estimator = RandomForestClassifier(), param_grid = param_grid, cv = 5, n_jobs = -1, scoring = 'accuracy')
```

```python
# Fit the gridSearch cv
grid_search.fit(X_train, y_train)
```

```python
# Best parameters
print(f'Best Parameters: {grid_search.best_params_}')
best_model = grid_search.best_estimator_
```

```python
# Save the best model
with open('Best_Random_Forest_Model.pkl', 'wb') as f:
    pickle.dump(best_model, f)

print("Model refinement completed and best model saved")
```
