# COPD in Nepal

### Factors that lead to COPD
1. Biomass fuel Exposure(Indoor air pollution)
2. Tobacco Smoking
3. Outdoor Air Polution
4. Occupational Exposure
5. Genetic Susceptibility
6. Respiratory Infections
7. Low Awareness and Helath Access
8. Climate and Environmental factors


### Signs and Symptoms of COPD
1. Chronic cough
2. Shortness to Breath(Dyspnea)
3. Excess Mucus production(sputum)(Phlegm)
4. Wheezing
5. Fatigue and low Energy
6. Chest tightness
7. Frequent Respiratory Infections
8. Cynosis(Bluish lips or fingernail beds)
9. Unintended Weight
10. Chronic Bronchitis and Emphysema Symptomps

```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
path = r"M:\Data Science\COPD_Prediction\Data\synthetic_COPD_data.csv"
df = pd.read_csv(path)
df.head()
```

```python
# Setup a large plot for better visibilty
plt.figure(figsize=(14,8))
```

```python
# Get info about the data
df.info()
```

## Univariate Analysis

```python
# What is the age distribution in my data
sns.histplot(df['Age'], kde= True, color = 'red')
plt.title("Age Distribution")
```

```python
# BMI Distribution
sns.histplot(df['BMI'],kde=True,color='blue')
plt.title("BMI Distribution")
```

```python
# Univariate analysis for the categorical features
# Gender distribution
sns.countplot(x = 'Gender', data=df,palette='viridis')
plt.title("Gender Distribution")
```

```python
# Smoking Status
sns.countplot(x = 'Smoking_Status', data = df, palette="Set1")
plt.title("Smoking Status Distribution")
```

### Bivariate Analysis

```python
plt.figure(figsize=(14,10))
```

```python
#1. Age vs COPD Diagonisis
sns.boxplot(x='COPD_Diagnosis', y= 'Age', data = df, palette='coolwarm')
plt.title("Age vs COPD Didagonisis")
plt.show()
```

```python
# BMI vs COPD Diagnosis
sns.boxplot(x='COPD_Diagnosis', y= 'BMI', data = df, palette='coolwarm')
plt.title("BMI vs COPD Diagnosis")
plt.show()
```

```python
# 3. Smoking Status
sns.boxplot(x='Smoking_Status', hue='COPD_Diagnosis', data = df, palette='Set2')
plt.title("Smoking Status vs COPD_Diagnosis")
plt.show()
```

```python
sns.countplot(x = 'COPD_Diagnosis',data= df, palette='viridis')
plt.title("Count Plot of COPD Diagnosis")
plt.show()
```

```python
# gender vs COPD diagnosis
sns.countplot(x='Gender', hue = 'COPD_Diagnosis', data = df, palette='Set2')
plt.title("Gender Vs COPD diagnosis count")
plt.show()
```

```python
# Correlation Matrix
data_corr = df[['Age', 'Biomass_Fuel_Exposure', 'Occupational_Exposure', 'Family_History_COPD', 'BMI', 'Air_Pollution_Level', 'Respiratory_Infections_Childhood', 'COPD_Diagnosis']]
corr = data_corr.corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# pari plot
sns.pairplot(df, hue = 'COPD_Diagnosis', diag_kind = 'kde', palette = 'Set1')
plt.title('Pair Plot')
plt.show()
```
